```python
# Extract all Wine ID

headers = {
  'Accept': 'application/json',
  'Authorization': 'Token 654181698a2795632e508be926ce4c021f431418'
}
request = Request('https://api.globalwinescore.com/globalwinescores/latest/?color=red&limit=10000', headers=headers)

response_body = urlopen(request).read()
print(response_body)
fulldata_red=json.loads(response_body)
json_normalize(fulldata_red['results'])

```

```python
#Create Dataframe Red
red=pd.DataFrame(json_normalize(fulldata_red['results']))
red.describe()
red.head(5)
```

```python
# Exam Data
red.country.unique()
red.color.unique()
```

```python
#Get White Wine
headers = {
  'Accept': 'application/json',
  'Authorization': 'Token 654181698a2795632e508be926ce4c021f431418'
}
request = Request('https://api.globalwinescore.com/globalwinescores/latest/?color=white&limit=10000', headers=headers)

response_body2 = urlopen(request).read()
# print(response_body2)
fulldata_white=json.loads(response_body2)
white=pd.DataFrame(json_normalize(fulldata_white['results']))
white.describe()
white.color.unique()
```

```python
#Get Pink Wine
headers = {
  'Accept': 'application/json',
  'Authorization': 'Token 654181698a2795632e508be926ce4c021f431418'
}
request = Request('https://api.globalwinescore.com/globalwinescores/latest/?color=pink&limit=10000', headers=headers)

response_body3= urlopen(request).read()
# print(response_body2)
fulldata_pink=json.loads(response_body3)
pink=pd.DataFrame(json_normalize(fulldata_pink['results']))
pink.describe()
pink.color.unique()
```
